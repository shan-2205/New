# Lecture 1 : The Agent Loop — How Codex Thinks

```python
'''

"Have you ever wondered —

    'How does Codex actually FIGURE OUT what to do ?'
    'Does it just guess ? 
     Does it have a brain ?'
    'Why does Codex sometimes ASK questions and sometimes just ACT ?'

These are excellent questions.
And the answer comes down to ONE concept —

    👉 The Agent Loop.


Once you understand this Agent loop,You will see exactly what is happening behind the screen.




========================================
PART 1 — The 4 Steps of the Agent Loop
========================================


Every time you give Codex a task,
it goes through these 4 steps —
AGAIN and AGAIN — until the task is DONE.


        ┌─────────────────────────────────┐
        │                                 │
        │   1️⃣  PLAN                       │
        │       ↓                          │
        │   2️⃣  DO                          │
        │       ↓                          │
        │   3️⃣  OBSERVE                     │
        │       ↓                          │
        │   4️⃣  NEXT STEP  →  back to PLAN  │
        │                                 │
        └─────────────────────────────────┘


Let me explain each step in plain English.


1️⃣ PLAN

    Codex first asks itself —
        'Okay, what does the user actually want ?'
        'What files do I need to look at ?'
        'What is my first move ?'

    It writes a small mental plan — like a to-do list.


2️⃣ DO

    Codex picks ONE step from the plan and DOES it.
        👉 Maybe read a file.
        👉 Maybe edit a file.
        👉 Maybe run a terminal command.

    Just one action at a time.


3️⃣ OBSERVE

    Codex looks at the RESULT.
        👉 Did the file open ?
        👉 Did the test pass ?
        👉 Did an error pop up ?

    This is the most important step. Codex LEARNS from the outcome.


4️⃣ NEXT STEP

    Now Codex decides —
        👉 'Cool, that worked. What is the next step ?'
        👉 'If anything failed. Let me re-plan.'
        👉 IF Codex is unsure. CODEX will ASK the user first.'

    And then it goes back to step 1 — PLAN — again.

    The loop runs over and over until the task is fully DONE.



Now lets understand how Agents works via REal -life example !
========================================
PART 2 — A Real-Life Example
========================================


Imagine you tell Codex —

👉 "Analyze my sales data and tell me why revenue dropped last month."

Here is what happens inside the loop :


    ROUND 1 :

    PLAN → "First, I need to understand the dataset."
    DO → Reads sales.csv
    OBSERVE → "Dataset has Date, Product, Revenue, Region columns."
    NEXT → "Let me check overall sales trend."



    ROUND 2 :

    PLAN → "Analyze monthly revenue."
    DO → Groups data by month
    OBSERVE → "Revenue dropped in March."
    NEXT → "Now I need to find which product caused this drop."



    ROUND 3 :

    PLAN → "Check product-wise sales."
    DO → Groups data by Product
    OBSERVE → "Product B sales dropped significantly."
    NEXT → "Let me check region-wise performance for Product B."





    FINAL OUTPUT :

    👉 "Revenue dropped in March mainly due to a price increase of Product B in the North region, which reduced sales."

And THIS is what you see happening behind the scenes
when Codex is working on your data.

Each step = one loop iteration.

Each line you see — is one step of this loop.



========================================
FINAL TAKEAWAY
========================================

    Whether it’s coding or Data analysis or any development tasks !
    Codex always works like this:

    👉 PLAN → DO → OBSERVE → NEXT STEP

    If you understand this 4-step rhythm —
        👉 You will write better prompts (bcz you know what Codex needs to plan).
        👉 You will be patient when Codex is mid-task (it is observing).
        👉 You will appreciate it when Codex asks you a question (smart agents ask).


    In the next lecture — let's talk about the BRAIN behind this loop.
    Bcz Codex doesn't just have one brain. It has multiple — called GPT-5 models —
    and each one is good at a different kind of work.

    See you in Lecture 2 !"

'''
```

# Lecture 3 : Reasoning Levels — Low, Medium, High

```python
'''

"Hi there ,  so far you have seen —

    👉 The agent loop  ( Codex's rhythm )
    👉 The model       ( Codex's brain )

Now we add ONE more dial —
    👉 Reasoning Level — which tells how DEEPLY the brain thinks before responding.

Even with the same model, you can tell Codex —
    'Think a little.'
    'Think medium.'
    'think hard.'

And the answer changes accordingly.


========================================
PART 1 — A Simple Real-Life Example First
========================================


Imagine you ask a friend two questions —

    Q1 :   'What is 10 * 9 ?'
           👉 They answer instantly. No deep thinking needed.

    Q2 :   'Should I quit my job and start a business ?'
           👉 They sit. They think. They take 10 minutes before answering.
           👉 Bcz the question deserves more thought.

    Same person. Same brain.
    Different DEPTH of thinking.

    Similarly , Reasoning level in Codex works exactly the same way..




========================================
PART 2 — The 4 Reasoning Levels
========================================


Codex gives you 4 reasoning levels —


🟢 LOW    →  thinks quickly, answers fast

    👉 Less time spent planning.
    👉 Best for SMALL, OBVIOUS tasks.
    👉 May miss edge cases on complex tasks.

    Use for :
        🔹 Rename a variable
        🔹 Fix a typo in code
        🔹 Add a single line of comment
        🔹 Quick formatting changes


🟡 MEDIUM ( default )  →  balanced thinking

    👉 Good planning + good speed
    👉 Best for MOST everyday tasks.
    👉 Reliable + reasonable speed.

    Use for :
        🔹 Add a new feature
        🔹 Fix a normal bug
        🔹 Refactor a single file
        🔹 Write a small function


🔴 HIGH   →  deep thinking, slower, very thorough

    👉 Maximum planning before execution
    👉 Handles edge cases pretty well
    👉 Slower — but the most reliable.

    Use for :
        🔹 Refactor across multiple files
        🔹 Logic-heavy tasks
        🔹 System design questions
        🔹 Anything where 'wrong answer = expensive'


🟣 EXTRA HIGH → maximum reasoning power

    👉 Deepest level of thinking
    👉 Very thorough, but slowest
    👉 Use ONLY when accuracy is critical

Use for :

    🔹 Complex architecture decisions
    🔹 Critical production bugs
    🔹 High-stakes code (finance, security, infra)
    🔹 When HIGH reasoning still gives weak results




========================================
PART 3 — The Beginner's Golden Rule of Reasoning 
========================================

    👉 Always START at MEDIUM level 
            MEDIUM is like the sweet spot for 80% of tasks.
            
    👉 DROP to LOW for trivial things.
            else your code quality will suffer and you will burn more time fixing mistakes.
            
    👉 BUMP to HIGH when results disappoint.
            else you will waste time  , money , tokens &  quota for no reason.
            
    👉 Use EXTRA HIGH sparingly
        → Only when:
        mistakes are costly
        problem is very complex
        or HIGH reasoning still fails






========================================
PART 4 — How to Change Reasoning Level
========================================


Two easy ways —


👉 INSIDE THE IDE EXTENSION :

    Look near the chat input — there is a small button
    that says  Low  /  Medium  /  High.
    Click it to switch.


👉 INSIDE THE CLI :

    During a session, just type :

        /reasoning

    Codex will show the 3 options. Pick one.
 


Quick tip —
    👉 You can change the level mid-conversation.
    👉 If Codex gave a weak answer at MEDIUM, switch to HIGH and rerun the prompt.



See you there !"

'''
```

# Lecture 5 : Approval Modes — Read-Only, Auto, Full Access

```python
'''


See We have already covered :
    ✅ The loop ( Agent Loop)
    ✅ The brain ( GPT Models by Codex )
    ✅ The depth of thinking ( Reasoning levels)

Now lets understand the last question —
    👉 How much PERMISSION does Codex have to actually DO things ?

See we know that Codex can read, edit files and run commands !
But you control HOW MUCH it is allowed to do without asking you first.

That control is called — Approval Mode.

If you to your VSCODE , u will see these approval options  , lets understand what they are & 
how they are useful !


    SEE didn't we already cover Sandbox vs Full Access in Section 2 ?

Yes — but that was about WHERE Codex can act .. ( like whether at your Project folder or it has access to your enture system)


But This lecture is about WHEN Codex needs your permission to act.
Let me make it crystal clear.

In simple terms : 


========================================
PART 1 — The Difference in One Line
========================================
    👉 Sandbox  =  WHERE Codex can touch. 
    👉 Sandbox keeps Codex inside a safe folder.
    
    👉 Approval makes Codex ask before risky steps.
    👉 Approval = means  WHEN Codex must ask.     ( permission )




& by Combining both ideas , and you have a very safe agent.





========================================
PART 2 — The 3 Approval Modes
========================================


Codex has 3 approval modes — same idea in BOTH the IDE and CLI.


🟢 MODE 1 — DEFAULT PERMISSIONS  ( also called CHAT mode )

    👉 Codex CAN read your files.
    👉 Codex CANNOT edit anything , cant run command !
    👉 It can only TALK and EXPLAIN.



🟡 MODE 2 — AUTO-REVIEW ( also called AGENT mode )
        In this mode , 
    👉 Codex can read, edit, and run — INSIDE your project folder.
    👉 If Codex wants to do anything OUTSIDE the folder, it ASKS you first.
    👉 If Codex wants to access the internet, it ASKS you first.

    👉 This is your GO-TO mode



🔴 MODE 3 — FULL ACCESS  ( also called AGENT FULL ACCESS )

    👉 Codex can read, edit, run — ANYWHERE on your machine.
    👉 Codex does NOT ask before any action.
    👉 Codex can also access the internet without approval.

    👉 Its Like giving full control — it does everything without asking, so use only when you fully trust it.

    Use for :
        🔹 Big migrations across many folders
        🔹 Long-running scripts where you cannot keep approving
        




========================================
PART 3 — The Perfect Choice : 
========================================


please listen carefully.

    👉 Stay in AUTO-review mode for everything in this course.
    👉 Use READ-ONLY only for code reviews / understanding code.
    👉 Use FULL ACCESS only after you have used Codex for weeks
       and know exactly what you are doing.





========================================
PART 4 — How to Change Approval Mode
========================================


Two easy ways —


👉 INSIDE THE IDE EXTENSION :

    Look near the chat input.
    There is a small dropdown that says  Agent.
    Click it. You will see —
        🔹 Default-Permissions (Chat Mode)
        🔹 Auto-Review Mode (Agent Mode)   ← default
        🔹 Full Access ( Agent Full Access Mode )
    Pick one. Done.




Quick tip —
    👉 You can switch modes ANYTIME — even mid-task.
    👉 So start in AUTO, and bump to FULL only IF the task really needs it.
    👉 Then drop back to AUTO when done.





Section 3 is officially done 🎉

So You now KNOW —
    👉 How Codex thinks
    👉 Which brain it uses
    👉 How hard it thinks
    👉 What it can act on

That puts you ahead of 90% of beginners using Codex today.


In Section 4 — we move from theory to PRACTICE.
We learn the K.I.M. Prompt Framework —

This is the section where things really start clicking.
See you in Section 4 !"

'''
```
