```python
'''

        Bad Prompt:
        “Analyze this data.”

        Convert this into Good Prompt : 
        
        GOOD PROMPT = Context + Task + Questions + Output Style



Simple Formula for Writing Better Prompts

1. Tell AI WHAT data you have [Context]
        Example :
        “I uploaded a stock market dataset.”
        This gives context to AI.



2. Tell AI WHAT TASK to perform [Task]
        Example:
        “give me a beginner-friendly summary.”


        This tells AI the main job.


 
3. Ask SPECIFIC QUESTIONS [Questions]
        “Analyze this data -
            Include : "

        Specific questions = better answers.



4. Tell AI HOW to explain [Output Style]

        Example:
        “Explain in simple language.”
            OR
        “Explain in 1 line each.”
        This controls the output style.

'''
```
