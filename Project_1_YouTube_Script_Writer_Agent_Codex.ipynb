# Lecture 5 : Chained Prompts + Memory Basics

```python
'''

"Okay — this is the most important lecture of this entire project.



We are going to learn TWO big things:
    1. Chained Prompts
    2. No Memory vs Memory (very important concept in AI agents)

Let's start.


========================================
PART 1 — What is a Chained Prompt?
========================================


See , what Most People do is : They write one big prompt like this:

    'Write me a YouTube script on AI jobs.'

And what do they get?
    👉 A generic, boring, mixed-up script.

Why?

Because AI gets overwhelmed when you ask for too many things at once.
Just like a human.



But CHAINED PROMPTING fixes this.

    Instead of ONE huge prompt…
    We break it into a CHAIN of smaller prompts.
    Each one focuses on just ONE job.
    


For our YouTube Script Writer Agent ,
the chain looks like this —


        ┌──────────────┐      ┌─────────────┐      ┌──────────────┐
        │  1. RESEARCH │ ───▶ │  2. OUTLINE │ ───▶ │  3. SCRIPT   │
        └──────────────┘      └─────────────┘      └──────────────┘
                │                    │                    │
                ▼                    ▼                    ▼
       research_notes.md       outline.md         final_script.md




Here is our chain:

    🔗 Prompt 1 : RESEARCH
        'Research the topic and save notes'
        → Output: research/research_notes.md
        
        👉 The output of step 1 becomes the input of step 2.
    
    
    🔗 Prompt 2 : OUTLINE
        'Read the research file and build an outline'
        → Output: outlines/outline.md
        👉 The output of step 2 becomes the input of step 3.


    🔗 Prompt 3 : SCRIPT
        'Read the outline and write the full script'
        → Output: scripts/final_script.md








Lets talk about "No Memory vs Memory"
========================================
PART 2 — No Memory vs Memory
========================================


Most beginners think —
    👉 'The agent remembers everything I said.'

Wrong.

By default , the GPT brain inside Codex is STATELESS.

    👉 No memory between sessions.
    👉 Every session/chat starts blank.



Now — how do we FIX this?
We ADD memory manually. 
There are Two ways to do that : 


👉 MEMORY TYPE 1 : AGENTS.md  (project memory)

    Anything we write inside AGENTS.md is memory.
    Claude reads it every time it starts.

    So in our project —
        Role, workflow, output rules…
        All of this is PERMANENT memory for this project.

    That's why AGENTS.md is so powerful.



👉 MEMORY TYPE 2 : FILES (task memory)


    The agent writes intermediate results to files —
        👉 research_notes.md
        👉 outline.md

    Then , when the agent moves to the next stage ,
    it READS that file before writing the next one.

    The file IS the memory.
    



Why does this matter?

    👉 If our agent had NO MEMORY —
        We would have to paste the research into every new prompt.
        We would have to re-explain everything & it is Super annoying.


    👉 With MEMORY (CLAUDE.md + Files[research/, outlines/, scripts/]) —
        The research flows automatically into the outline.
        The outline flows automatically into the script.
        Claude 'remembers' without remembering.


So the rule is:

    There is No memory by default.
    You can Add memory in Projects by using AGENTS.md (rules) + Files (data).
    You can Chain your prompts through files..
    
    & This is how you build REAL agents..
    




    


In the next lecture, we finally run all 3 chained prompts
and watch our agent write a full YouTube script from scratch.

Get ready — this is where the magic happens."

'''
```

# Lecture 7 : Making It Reusable — The `/script` Slash Command

```python
'''

"Welcome to the lecture that turns this project from a NICE agent
into a REUSABLE TOOL that you will keep using for years.

Right now , every new video looks like this —

    👉 Open Codex.
    👉 Type Prompt 1   ( long )
    👉 Wait.
    👉 Type Prompt 2   ( long )
    👉 Wait.
    👉 Type Prompt 3   ( long )

That works. But that is the LIFE of a manual user , not an agent owner.

Let's fix it with one of the most underrated features in Codex —
custom SLASH COMMANDS.


========================================
PART 1 — Quick Refresher : What Are Slash Commands ?
========================================

    👉 Slash commands are shortcuts you type inside Codex.
    👉 They start with a /.
    👉 Codex ships with built-in ones —   /init , /status , /model , /approvals , /quit.

But here is the gold —

    👉 You can CREATE YOUR OWN slash commands.

    Whatever workflow YOU run again and again —
        👉 Bundle it once.
        👉 Save it as a file.
        👉 Type one slash command forever.


For our agent , we will create —

    /script

Whenever you type    /script  
                            topic="..."  
                            audience="..."
the entire 3-stage pipeline runs automatically.


========================================
PART 2 — Where Custom Slash Commands Live
========================================


Codex looks for custom commands inside a special folder
in your project —

        .codex/
          └── prompts/
                └── script.md


Notice 3 things —

    1. There is a DOT in front of  .codex.   ( hidden folder )
    2. Each Markdown file inside becomes one slash command.
       File name = command name.    ( script.md  →  /script )



========================================
PART 3 — Step-by-Step : Build the /script Command
========================================


🔹 Step 1 — Create the hidden folder

    Inside  youtube_script_agent/  create a new folder — .codex
    Then , inside  .codex  create another sub-folder — prompts

    Final path —    youtube_script_agent/.codex/prompts/



🔹 Step 2 — Create the command file

    Inside  .codex/prompts/  create a new file called — script.md
    Open it.


🔹 Step 3 — Paste the bundled 3-stage prompt

    Copy this into  script.md  exactly :


    # /script — Full YouTube Script Pipeline

    You are the YouTube Script Writer Agent defined in AGENTS.md.

    Inputs you will receive in the user message —
        - topic    : the YouTube video topic
        - audience : who the video is for

    Run all 3 stages in order. Do NOT stop between stages.


    ## Stage 1 — RESEARCH
    Use the Web tool. Find 5 to 7 fresh sources from the LAST 6 MONTHS.
    For each source , note — Title , URL , 2 useful plain-English facts.
    Save to    research/research_notes.md

    ## Stage 2 — OUTLINE
    Read    research/research_notes.md.
    Build a 7-minute YouTube outline —
        HOOK ( under 10s , curiosity driven )
        3 MAIN POINTS ( each with a 1-line takeaway )
        CTA ( like , subscribe , comment )
    Save to    outlines/outline.md

    ## Stage 3 — FINAL SCRIPT
    Read    outlines/outline.md.
    Write the full spoken script —
        spoken tone , beginner words , short paragraphs ,
        strong hook in first 10s , strong CTA at the end.
    Save to    scripts/final_script.md

    ## Final Step
    Print a summary —
        ✅ research_notes.md saved
        ✅ outline.md saved
        ✅ final_script.md saved
    Then stop.


    Save the file.


🔹 Step 4 — Restart Codex ( important ! )

    Close the Codex session.
    Re-open Codex inside the same folder.

    Codex picks up new slash commands ONLY on a fresh start.


🔹 Step 5 — Use it !

    Inside Codex , type :

        /script  topic="AI agents for beginners"  audience="complete beginners"

    Press Enter.

    Sit back.

    Codex now —
        1. Runs research.   ( saves research_notes.md   )
        2. Runs outline.     ( saves outline.md          )
        3. Writes script.    ( saves final_script.md     )
        4. Prints the ✅  summary.

    All from ONE command.


========================================
PART 4 — Why This Is Such a Big Deal
========================================

🔹 Before the slash command —

    👉 You typed 3 long prompts every video.
    👉 You had to remember the exact wording.
    👉 You had to remember the exact filenames.
    👉 If you forgot one detail , the chain broke.


🔹 After the slash command —

    👉 You type ONE word :    /script  +  👉 Plus 2 inputs :  topic + audience
    👉 You can hand this folder to a teammate
       and they get the SAME results.


This is how solo creators turn into agentic studios.
Not by working faster.
By DELEGATING faster.




Alright — your YouTube Script Writer Agent is COMPLETE.
You have —

    ✅  A clean folder structure
    ✅  An AGENTS.md brain
    ✅  Live web access
    ✅  A 3-stage chain with file-based memory
    ✅  A one-word reusable command  →  /script

In the last lecture , we will quickly recap the 5 mental models
that this ONE project just taught you for free.

See you there !"

'''
```
