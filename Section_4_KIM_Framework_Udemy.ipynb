# Lecture 2 : K = Knowing Your Goal

```python
'''

"Welcome to the FIRST letter of the framework — K.

K stands for —   Knowing your Goal.


========================================
PART 1 — What Does 'Goal' Actually Mean ?
========================================


When I say goal — 
I mean —
    👉 The OUTCOME you want must be written in ONE sentence.

        Build  X  that does for Y.
                OR
        what we are actually building ?

That's it !'
X = the thing.
Y = what the thing does.

eg


❌ WITHOUT a goal

    Prompt : 'Add login.'

    Codex thinks — login to what ? or For whom we are building this Login page ?
    It guesses. 
    You suffer.


✅ But when u add a clear goal like : 

    'Build a login page that lets users sign in with Google.'
     Now Codex knows —
        👉 X  =  a login page
        👉 Y  =  lets users sign in with Google


    'Build a script that reads a CSV and prints total sales by month.'
     Now Codex knows —
        👉 X  =  script
        👉 Y  =  that reads a CSV and prints total sales by month.


    So always follow this simple The 1-Sentence Test (Build  X  that does for Y.)





========================================
PART 2 — Quick Practice ( Do This Now )
========================================


Pause the video for a moment !
Take any task & try to write it using The 1-Sentence Test :   Build X that does Y.


Some beginner-friendly examples —

    👉 Build a Python script that renames all .jpg files in a folder.
    👉 Build a function that converts USD to INR using a public API.
    👉 Build a small webpage that shows today's weather for any city.
    👉 Build a Markdown file that summarizes my README in 5 bullet points.

    [Open VSCode , for this CSV file , say :
    
    👉 Build a Python script that read csv file ]

========================================

========================================


You have nailed the K of K.I.M.

Now in the next lecture , we will talk about "I" - > Including the right Context.

Bcz Without context — 
    Codex will hallucinate.
    It will invent files that do not exist.
    It will assume a stack you are not using.

So in the next lecture — we attach the second letter : "I"



See you in the next lecture !"

'''
```

# Lecture 3 : I = Including the Right Context

```python
'''

"Now we are at the SECOND letter — I.

I stands for —   Including the right Context.

If K told Codex WHAT to do or What Goal is  —
then 
I tells Codex WHERE to look.


And without I —  
Codex starts inventing/hallucinating things. 
Guessing imports. 

Let me show you why context is the secret weapon !
Lets connect this idea with a real life example !



========================================
PART 1 — Why Context Matters So Much
========================================


Imagine you walk into your office on Day 1.
Your boss says —

    'Fix the bug in the report.'

You stand there blinking.
    👉 Which report ?
    👉 Which file ?
    👉 Which folder ?
    👉 Where is the codebase even ?

You have no context. 
You will fail no matter how smart you are.

Codex feels EXACTLY like that on Day 1.
It does not magically know your project.
It can SEE your files — but only if you point at them.


    The rule —
        👉 No context  =  Codex hallucinates.
        👉 Right context  =  Codex is highly-accurate.




========================================
PART 2 — The 3 Types of Context
========================================


When I say 'context', I mean 3 specific things —


    👉 Type 1 — FILE CONTEXT
        Which file(s) should Codex look at ?
        Which folder is relevant ?

    👉 Type 2 — STACK CONTEXT
        Which Programming language ? 
        Which framework ? 
        Which version of package should be considered ?

    👉 Type 3 — DOMAIN CONTEXT
        What is this project even about ?
        Is it a sales dashboard ? A blog ? A trading bot ?


Most of the time — File Context is the BIGGEST win.
Get that one right, and 80% of hallucinations will disappear !


# lets talk about some Tricks which can use to mention context ! 

========================================
PART 3 — Trick #1 : The @file Shortcut
========================================

[VSCODE]

Instead of typing —
    'Look at the file at src/utils/payment.py and fix...'

You just type —
    '@payment.py — fix the bug in this Python file 




========================================
PART 4 — Trick #2 : Drag and Drop ( IDE only )
========================================


If you are using Codex inside IDE 
you can literally DRAG a file from the file explorer
and DROP it into the chat box.

    👉 Drag        →   src/utils/payment.py
    👉 Drop into   →   the Codex chat
    👉 Codex sees  →   the full file as context




========================================
PART 5 — Trick #3 : Tell The Right Stack
========================================


Even with files attached — although its a good practise to
tell Codex about your STACK.

Bcz the same task in Python is different from the same task in Node.


❌ Without stack context

    'Build a script that fetches data from an API.'

    Codex picks — 
    could be Python ? 
    Could be Node ? 
    Random selection 


✅ With stack context

    'Build a Python script (using requests library)
     that fetches data from this API.'

     Codex now picks — Python + requests.






Now lets have a mini DEmo 
========================================
PART 6 — Mini Demo
========================================

[Open Vs Code]
Let's contrast two prompts that share the SAME goal.


❌ Without I

    👉 Build a Python script that read csv file 

    Codex thinks — 
        Which file ? 
        Which language ?
        Maybe finds it. Maybe not.


✅ With I

    '@sales.csv — mention your CSV file !
      👉 Build a Python script that reads this @sales.csv file
      
    Codex thinks —
        File is @sales.csv
        Goes straight to folder & find this file..




========================================
WHAT'S NEXT
========================================


So far , we have covered Two letters..

    👉 K  ✅  Knowing your Goal
    👉 I  ✅  Including the right Context
    👉 M  ⏳  Managing Constraints


Bcz this is where you tell Codex —
    👉 Don't touch this file.
    👉 Use this naming style.


Let's go nail M in the next lecture !"

'''
```

# Lecture 4 : M = Managing Constraints  !

```python
'''

"Welcome to the THIRD and final letter — M.

M stands for —   Managing Constraints.


========================================
PART 1 — What Are Constraints ?
========================================


Constraints are RULES of the game.
They tell Codex —
    👉 What it MUST do
    👉 What it MUST NOT do
    👉 When it is FINISHED

Without constraints — 
    It refactors stuff you did not ask. 
    It adds libraries you do not need.
    It edits 10 files when only 2 needed change.


With constraints — 
    Codex stays in its lane. 
    Faster Code reviews.



Now , lets talk about Types of Constraints
========================================
PART 2 — The 3 Types of Constraints
========================================


Type A — TECH STACK CONSTRAINT
        Here u can mention TECH of your Project like :
        
        [Open VSCode ]
         👉 Build a Python script(using Pandas not raw csv module) that reads this @sales.csv file


Type B — STYLE / NAMING CONSTRAINT
        Here u can mention NAMING CONSTRAINT of your Project like :
        
        [Open VSCode ]
        👉 Build a Python script using Pandas that reads this @sales.csv file

        Constraints:

        Use df_sales as the variable name
        Keep each function under 30 lines
        Add a docstring to every function
    



Type C — DON'T-TOUCH CONSTRAINT  ( my favorite )
        While working on any Project , if u have sensitive info :
        U can mention here !
        
        [Open VSCode ]
        👉 Build a Python script using Pandas that reads this @sales.csv file

        Constraints:

        Use df_sales as the variable name
        Keep each function under 30 lines
        Add a docstring to every function
        Do not modify any existing files
        Do not install or update any packages
        'Don't change anything outside this folder 





'''
```

# Lecture 5 : 'Done When test' !

```python
'''

========================================
PART 3 — The SECRET WEAPON : 'Done When test'
========================================


Lets talk about   Done-When  test.

    👉 You add ONE line at the end of your prompt :
        'You are done when ___________.'

And here ___________ ,  u can mention a TESTABLE condition.


Examples —
    👉 'You are done when the homepage of your Website loads in under 1 second.'
    👉 'You are done when the CSV file produces a chart with 12 monthly bars.'



Why is this so powerful ?

Bcz Codex is an AGENT.
Agents loop. They try, check, retry.

    👉 Without 'done when' — 
        Codex or any AI AGent may stops randomly.
        It thinks , it'looks good' and ends.

    👉 But With 'done when' — 
        Codex keeps iterating  , until the condition is TRUE.
        So It does not stop until your condition is met.


        [VS Code]
        👉 Build a Python script using Pandas that reads this @sales.csv file

        Constraints:

        Use df_sales as the variable name
        Keep each function under 30 lines
        Add a docstring to every function
        Do not modify any existing files
        Do not install or update any packages

        Done when:
        The script successfully reads @sales.csv into df_sales and prints the first 5 rows



Before pressing Enter, always check KIM framework : 

    K  →  Did I write 'Build X that does Y' ?
    I  →  Did I attach the right files / mention the stack ?
    M  →  Did I add at least ONE 'don't touch' or one 'done when' line ?

If all three are YES — your prompt is Codex-ready or "agent-ready"
Press Enter with confidence.








========================================
WHAT'S NEXT
========================================


You have officially learned the entire K.I.M. framework.
    👉 K  ✅  Knowing your Goal
    👉 I  ✅  Including the right Context
    👉 M  ✅  Managing Constraints + Done-When

    👉 ✅ Parallely we have done hands-on as well !

So in the next lecture — I will take a real,
messy, beginner prompt — the kind you and I both write daily —
and rewrite it LIVE using K.I.M.

You will see the BEFORE and the AFTER effect !


Let's go !"


'''
```

# Lecture 6 : K.I.M. Live Demo — Rewriting a Vague Prompt

```python
'''

"Alright — this is the lecture you were waiting for.

We have learned the Entire "KIM" framework 
Now let's have some more hands-on !


I will take ONE real bad prompt that most of the beginners uses
& then I will Show you what Codex does with it.

Then rewrite it using K.I.M. framework and run it again.
& then we will show you the diff !



========================================
PART 1 — The Setup
========================================


So this is a small project on my laptop..
You can download it from resources !
It is a simple Python sales tracker —
    👉 sales_data.csv     ( 12 months of sales numbers )
    👉 utils/loader.py    ( reads the CSV )
    👉 main.py            ( prints some output )

I want Codex to add a new feature —
a chart that shows monthly sales.

Now — let's see what happens when I write a LAZY prompt.
Then a K.I.M. prompt.


========================================
PART 2 — Without K.I.M : vague, lazy Prompt 
========================================


I open Codex inside IDE 
I type :
        Add a chart to my project.

Press Enter.


Watch what Codex does —
    👀 'Looking at the project structure...'
    📖 'Reading main.py ...'
    📖 'Reading utils/loader.py ...'
    🤔 'Need a charting library — installing matplotlib AND seaborn AND plotly...'
    ✍️'Creating chart_generator.py ...'
    ✍️'Creating dashboard.html ...'
    ✍️'Editing main.py ...'
    ✍️'Editing requirements.txt ...'

After 2 minutes — Codex finishes.

What did I get ?
    👉 3 charting libraries installed ( I needed 1 )
    👉 An HTML dashboard ( I never asked for that )
    👉 main.py rewritten ( I did not want it touched )
    👉 The chart… is pie chart of unknown data

Total mess.




========================================
PART 3 — With K.I.M. : 
========================================

Now lets rewrite the prompt using K.I.M.


K  →  Knowing the Goal

    'Build a Python script that generates a bar chart
     of monthly sales from sales_data.csv.'


I  →  Including the Context

    '@sales_data.csv
     @utils/loader.py — use the existing loader function
     to read the CSV.'


M  →  Managing Constraints

    '👉 Use matplotlib only. Do NOT install seaborn or plotly.
     👉 Save the chart as monthly_sales.png in the project root.
     👉 Do NOT modify main.py or any existing file.
     👉 You are done when monthly_sales.png exists
        and shows 12 bars — one per month.'


Combined into one prompt —

        Build a Python script that generates a bar chart of
        monthly sales from sales_data.csv.

        Context :
          @sales_data.csv
          @utils/loader.py — use the existing loader function.

        Constraints :
          - Use matplotlib only. No other libraries.
          - Save chart as monthly_sales.png in the project root.
          - Do NOT modify main.py or any existing file.
          - You are done when monthly_sales.png exists with 12 bars.

Press Enter.


Watch what Codex does THIS time —
    👀 'Reading sales_data.csv ...'
    📖 'Reading utils/loader.py ...'
    ✍️'Creating chart.py — using matplotlib only ...'
    ✍️'Running chart.py ...'
    ✅ 'monthly_sales.png created with 12 bars.'
    ✅ 'Done condition met.'

Total time — about 30 seconds.

What did I get ?
    👉 ONE clean new file ( chart.py )
    👉 ONE chart image ( monthly_sales.png )
    👉 Zero unwanted changes
    👉 Codex stopped at the right moment

Same project. Same Codex. Same model.
Just a different PROMPT.


========================================
PART 4 — The Side-By-Side Comparison
========================================


Look at the two outcomes side by side —

    BEFORE  ( vague prompt )
        👉 4 files changed
        👉 3 libraries installed
        👉 Wrong chart type
        👉 Took more time in execution
        👉 Followed by 4 fix-up prompts

    AFTER  ( K.I.M. prompt )
        👉 1 file added
        👉 1 library used
        👉 Correct chart
        👉 Took 30 seconds
        👉 Zero fix-up prompts

This is not a small upgrade.
This is a 10x productivity unlock.

And ALL of it came from
KIM Framework !






'''
```
