# Lecture 1 : What AGENTS.md Is and Why It Matters

```python
'''

========================================
PART 1 — Why AGENTS.md Matters
========================================



So Imagine this…

let say : You are a Data Analyst at Uber & your Job is to Analyze Uber Ride data !

You spend 2 hours with Codex & built amazing thing !


With the help of Codex Agent ,You’ve done a lot:

    Collected Uber data from various sources
    Cleaned messy data
    Built KPIs (Total rides, Peak hours , Revenue per trip)
    Built dashboards [show dashbaord] so that u can showcase it to Business Team/managers 
    
    
    
    
    
Everything is working perfectly.
You’re confident.
You’re proud..





Now next day… Your Manager says to improve this dashbaord
Then You open a new AI session in Codex !
And you say:
    👉 “Help me improve my Uber analysis dashboard”

And suddenly…
The Codex asks:

    👉 “What does your dataset look like?”
    👉 “What columns are available?”
    👉 “How are you calculating revenue?”


And you’re like…
😩 “I already explained all of this yesterday!”

So its Frustrating… right?

[Pause]



👉 Bcz LLMs like GPT(which codex uses internally) don’t have memory across sessions..
    LLMs are memory-less / stateless 
            OR
    GPT don’t remember anything across sessions.

They only remember what’s inside the current conversation/session..
And this creates a HUGE problem…






So the real question is…

👉 How do we make Codex REMEMBER our project?"
    One way is : Explain things from scratch into every new session ( thats practically impossibe 
                    in bigger projects)
                    
    Second way is : You could give Codex/AI one file… & that file contains all your
                    Project info 
                    And then  , Codex understands your entire project instantly 
                    & that file is known as AGENTS.md !





========================================
PART 2 — A Simple One-Line Definition
========================================


AGENTS.md is :

    👉 A plain text file in your project folder which contains your project's rules, conventions, and warnings.
    👉 This file is Read by Codex BEFORE every single task it does in your project.


In simple words —

    👉 README.md  is designed for humans which tells what your project does..
    👉 AGENTS.md  is designed for Codex which tells to your Codex Agent what your project is all about 


    👉 README.md  =  for humans  ( what your project does )
    👉 AGENTS.md  =  for Codex   ( how YOUR project should be worked on )




========================================
PART 3 — Where Does this "AGENTS.md" Live ?
========================================


Super simple.

    👉 It lives in the ROOT of your project folder.
    👉 The filename is exactly :   AGENTS.md



Example : How folder structure for our project looks like :

    📂 Sales_data_analysis/
        ├── 📜 AGENTS.md             ← this is the magic file
        ├── 📜 README.md             ← for humans
        ├
        ├── 📂 data/
        │     └── sales.csv          ( product_name, quantity, Price, region )
        └── 
              


Codex sees AGENTS.md in the root → reads it → applies your rules.
Done.

Now in the next session , we will write First AGENTS.md file !
See you in Lecture 2 !"


'''
```

# Lecture 2 : Writing Your First AGENTS.md ( Sales Data Project )

```python
'''

[VSCODE]

"Now its Time to actually write our first AGENTS.md file.

We will continue with the same Project which is : 
    📂 Sales_data_analysis

It is a small Python data Analysis project that analyzes sales data.
The data has 4 columns —
    👉 product_name
    👉 quantity
    👉 Price            ( notice : capital P )
    👉 region
    

Open your project root.
Create a new file. Name it exactly :   AGENTS.md

Now We are going to teach Codex EVERYTHING about this project
in one small file "AGENTS.md"




Lets talk about 5 imp Sections that Matters for this AGENTS.md
========================================
PART 1 — The 5 Sections That Matter
========================================

So its a good practise to include these 5 sections !

    1️⃣  Project Overview         →  what is this project ?
    2️⃣  Tech Stack                →  what tools and libraries we use
    3️⃣  Folder Structure          →  where things live
    4️⃣  Code Style                →  naming, formatting, conventions
    5️⃣  Don't Touch Rules         →  files Codex must NEVER edit

so these are 5 imp ones !

Optionally you can add — testing rules, commit message style, etc.
But for beginners, these 5 are more than enough.


========================================
PART 2 — Let's Build Each Section, Step by Step
========================================

[VS CODE]
─────────────────────────────────────────
SECTION 1 — PROJECT OVERVIEW
─────────────────────────────────────────

    Goal : Tell Codex in 2-3 lines what this project does.


    # Sales Data Analysis

    ## Project Overview
    A Python project that analyzes a sales dataset.
    The dataset has 4 columns: product_name, quantity, Price, region.
    The goal is to produce reports like total sales per region,
    top-selling products, and average price by region.


✅ Why this works : bcz 
    👉 It Tells Codex the DOMAIN ( sales data )
    👉 It Lists the column names EXACTLY 
    👉 It Gives a sense of typical tasks Codex will be asked to do


─────────────────────────────────────────
SECTION 2 — TECH STACK
─────────────────────────────────────────

    Goal : It will Tell Codex which libraries to use AND which to avoid.


    ## Tech Stack
    - Language: Python 3.11
    - Data handling: Pandas (always prefer pandas over raw loops)
    - Plotting: Matplotlib (no other plotting libraries)
    - Tests: Pytest


✅ Why this works :
    👉 'always prefer pandas over raw loops'  →  Codex won't write a slow for-loop
    👉 'no other plotting libraries'           →  Codex won't randomly bring in seaborn
    👉  Specific Python version                 →  Codex uses correct syntax


─────────────────────────────────────────
SECTION 3 — FOLDER STRUCTURE
─────────────────────────────────────────

    Goal : Tell Codex where to put new code.
    Without this, Codex will create files in random places.


    ## Folder Structure
    - `src/`              : All Python source code
    - `data/`             : Raw CSV files
    - `tests/`            : Pytest test files (one test file per source file)
    - `reports/`          : Output charts and PDFs



─────────────────────────────────────────
SECTION 4 — CODE STYLE
─────────────────────────────────────────

    Goal : Tell Codex how YOUR code should look.
    Codex matches your style naturally — but only if you tell it.


    ## Code Style
    - Functions must have a one-line docstring explaining what they do.
    - Use the EXACT column names from the data: product_name, quantity, Price, region.
      Note: Price has a capital P. Do not lowercase it.
    - Keep functions short — under 25 lines if possible.




─────────────────────────────────────────
SECTION 5 — DON'T TOUCH RULES
─────────────────────────────────────────

    Goal : Protect critical files from Codex.
    Even smart agents can accidentally edit something they should not.


    ## Don't Touch
    - Never edit `data/sales.csv`. It is the raw source of truth.
    - Never delete or rename files in `reports/`. 
    - Do not add new dependencies without asking first.







Now in the next lecture — let's actually TEST our AGENTS.md.

We will ask Codex to add a feature to the Sales project
and watch — does it follow our rules ? Or does it break them ?

See you in Lecture 3 !"

'''
```

# Lecture 3 : Testing Your AGENTS.md

```python
'''

"People always make ONE big mistake with AGENTS.md.

They write the file —
and then ASSUME Codex is following it.

Don't assume. 
ALways TEST.


========================================
PART 1 — Testing AGENTS.md
========================================


Testing AGENTS.md is super simple. Just 3 steps :

    1️⃣  Give Codex a small NEW task in the project.
    2️⃣  Watch what it generates.
    3️⃣  Score it against your AGENTS.md rules.


If Codex follows 90%+ rules — your AGENTS.md is solid ✅
If it breaks rules — your AGENTS.md needs sharpening 🔧



========================================
PART 2 — Our Test Task
========================================
[VSCODE]

Open your Codex chat Panel in the VS CODE !
Make sure your AGENTS.md is saved at the root.

Now type this exact prompt :


    👉 'Add a function that returns the top 3 products
        by total revenue. Revenue = quantity × Price.
        Group by product_name. Return a list of 3 names.'


Press Enter and let Codex work.


========================================
PART 3 — The Scoring Checklist
========================================


When Codex finishes, do NOT just glance at the output.
Pull up your AGENTS.md side by side and SCORE it.

Here is your checklist for our project —


    ✅ Did it follow folder structure rule ?
        Did Codex put the function in  src/analysis.py  ?

    ✅  Did Codex use Pandas groupby — not raw Python loops ?

    ✅  Did Codex use the EXACT column name  "Price"  ?

    ✅  Did Codex avoid touching data/sales.csv ?


Count your score :

    4 / 4  →  AGENTS.md is excellent  🏆



What if Codex Breaks a Rule
========================================
PART 4 — When Codex Breaks a Rule — How to Iterate
========================================


Don't panic when Codex breaks a rule.
It happens. It is part of the process.

Here is the iteration loop —


🔁  THE AGENTS.md ITERATION LOOP :

        1️⃣  Spot the rule that was broken.
        2️⃣  Ask WHY — was the rule vague or missing ?
        3️⃣  REWRITE the rule to be more SPECIFIC.
        4️⃣  Save AGENTS.md.
        5️⃣  Run the SAME prompt again.
        6️⃣  Recheck the output.

Repeat until the rule is followed 95% of the time.




Now , lets do a Quick Sanity-Check !
========================================
PART 5 — A Quick Sanity-Check Trick
========================================


SO if you Want to confirm Codex is READING your AGENTS.md or not ?
Use this simple trick.

Just ask Codex —

    👉 'Before doing any task, list the top 5 rules
        from my AGENTS.md file in your own words.'

If Codex repeats your rules accurately —
    ✅ Your AGENTS.md is being read.

If Codex says 'I don't see any AGENTS.md file' —
    🚨 Either the file is in the wrong folder.
    🚨 Or named incorrectly  (must be exactly AGENTS.md, all caps).
    🚨 Or you are running Codex from the wrong directory.




========================================
PART 6 — Your Mini Exercise
========================================


Stop the video and try this —

    1. Open your Sales_data_analysis project ( with AGENTS.md saved )
    2. Ask :
         'Add a function that returns total revenue per region.
          Revenue = quantity × Price. Group by region.'
    3. Score the output using the 5-point checklist
    4. If anything fails — sharpen that rule and try again


just do it , & belive me it will teach you more about AGENTS.md
than reading 10 articles.


========================================
FINAL TAKEAWAY
========================================


Just remember —

    👉 Don't assume Codex is following your AGENTS.md. TEST it.
    👉 If a rule is broken — sharpen it. Don't blame Codex.
    👉 AGENTS.md is LIVING DOCUMENTATION. It improves with iteration.


Now in the final lecture of this section —
I want to share the 3 BIGGEST mistakes I see beginners make
when writing their first AGENTS.md.


See you in Lecture 4 !"

'''
```

# Lecture 4 : The 3 Mistakes Beginners Make

```python
'''

"HI there , in this lesson , we will discuss Top 3 mistakes one by one that learners make 
    while designing brain of Agent "AGENTS.md" file !


If you avoid these 3 mistakes —
your AGENTS.md will be better than 90% of the teams


Let's go through each one — with our Sales_data_analysis project
as the example.


========================================
MISTAKE 1 — TOO LONG  ( the 200+ line monster )
========================================


When beginners write their first AGENTS.md, they get excited.

They start adding —
    🔹 'Always be polite.'
    🔹 'Refer to the company style guide v3.4.'
    🔹 'Use modern best practices.'
    🔹 'Be helpful and clear.'

Soon — the file is 300 lines long. 


🚫 PROBLEM :
    👉 Codex has a limit on how much it can read from AGENTS.md.
    👉 If your file is too long, important rules get IGNORED.
    👉 more rules = more confusion = worse output.

Long does not mean better. Long means cluttered.


✅ THE FIX :
    👉 Aim for under 200 lines. Ideally 50-100 lines.
    👉 Keep ONLY the rules that matter for code quality.
  


Bad rule  ❌ :   'Write good code.'
Good rule ✅ :   'Use Pandas groupby for any aggregation across rows.'




========================================
MISTAKE 2 — VAGUE RULES  ( 'write good code' )
========================================


This is the BIGGEST one

Examples for our Sales project —


    🚫 Beginners write : "Write clean code."
    ✅ 'Functions must be under 25 lines and have a single purpose.'

    🚫 'Write good comments.'
    ✅ 'Every function must have a one-line docstring saying what it does.'

    🚫 'Handle errors well.'
    ✅ 'When the CSV is missing, raise FileNotFoundError with a clear message.'


See the difference ?
    👉 Vague  →  Codex guesses.
    👉 Specific  →  Codex follows.



========================================
MISTAKE 3 — FORGETTING TO UPDATE IT
========================================


This is the silent killer.

Beginners write AGENTS.md once on Day 1.
Save it. 


But Three months later —
    👉 The project has new folders.
    👉 The team switched from Matplotlib to Plotly.
    👉 The data has 4 columns now, in future it can be 6-7
 

SO you have to update your AGENTS.md file as well , else Codex starts following OUTDATED rules.

✅ THE FIX — treat AGENTS.md as LIVING documentation.

Three simple habits :

    1️⃣  Whenever your project changes, update AGENTS.md in the SAME commit.
         New folder ?  →  add it to AGENTS.md.
         New library ?  →  add it to AGENTS.md.
         New column ?  →  add it to AGENTS.md.

    2️⃣  Once a month — do a 5-minute review.
         Open AGENTS.md.
         Read each rule.
         Ask : is this still true ?
         Update / delete anything outdated.

    3️⃣  Whenever Codex breaks a rule MORE THAN ONCE —
        Go & update AGENTS.md.






Now in Section 6 — we level up.
We will learn the Codex Slash commands you will use every single day —
the workflow shortcuts that separate beginners from power users.

See you in Section 6 !"

'''
```
