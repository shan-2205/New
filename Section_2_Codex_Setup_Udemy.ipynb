# Lecture 4 : Sandbox vs Full Access — The Safety Toggle

```python
'''

"Okay — in this lecture

we will understand MOST IMPORTANT safety lesson in the entire course.

If you understand this ONE concept properly —
    👉 You will use Codex with full confidence.




========================================
PART 2 — The Two Modes
========================================

So there are 2 modes to access files & folder by your Agent "Codex" 


🟢 MODE 1 — SANDBOX MODE  ( the SAFE mode — toggle ON )

    👉 Codex can ONLY touch files inside the project folder.
    👉 Codex cannot touch files outside the project folder.
    👉 Codex cannot run dangerous shell commands.
    👉 Codex cannot delete random files on your computer.

    [show project on VSCODE]


🔴 MODE 2 — FULL ACCESS MODE  ( the OPEN mode — toggle OFF )

    👉 Codex can touch ANY file on your computer.
    👉 Codex can run ANY shell command.
    👉 Codex can edit or delete files OUTSIDE the project folder.
    👉 Codex can install software, change system settings, etc.

        In this mode , Codex has the access to your entire system !
        It is Powerful — but risky.








And here is my golden rule —

    👉 Stay in Sandbox mode 95% of the time.
    👉 Switch to Full Access ONLY when you really need to.
    👉 And most imp when your task is done — switch back to Sandbox.

    Most of the time — you will not need Full Access mode at all.
    Bcz 99% of coding tasks happen INSIDE your project folder.


========================================
PART 3 — When DO You Need Full Access ?
========================================


Honestly ? 
Very rarely as a beginner.
But there are 3 cases —

    1. Codex needs to install software globally on your machine.
    2. Codex needs to fetch something from the internet
       that the sandbox is blocking.
    3. Codex needs to edit a file OUTSIDE the project folder
    
    👉 NEVER use Full Access on a folder that contains :
        🚫 your real bank account scripts
        🚫 your customer database
        🚫 your secret API keys




========================================
PART 4 — How to Switch Modes
========================================


Here is the good news —
Codex has SAFE DEFAULTS already built in.

By default — you are already in the safe zone.
    👉 Codex asks you for permission before doing anything risky.
    👉 You see a small prompt — 'Approve ?  yes / no'
    👉 If you say no — Codex stops.


Inside the IDE extension, you have 3 buttons :

    🟢 Chat                     →  Read-only. Safest.
                                    Codex can talk but cannot touch any file.

    🟡 Agent  ( default )       →  Codex can read, edit, and run
                                    INSIDE the project folder only.
                                    Asks you before doing anything outside.

    🔴 Agent ( Full Access )    →  No fence. Codex does whatever it wants.
                                    Use ONLY when you really know what you are doing.


Inside the CLI — same idea, 
    U can check all the modes by command : /approvals

    👉 Read Only — Look but don't touch; every edit or internet request needs your approval.
    👉 Default — Free reign inside your project folder; anything outside it requires your go-ahead.
    👉 Auto-review — Same as Default, but a second AI approves the borderline calls instead of you.
    👉 Full Access — No fences, no asking; Codex does whatever it wants, anywhere on your machine.



                                           
                                                


========================================



Now — finally — let's run our first real Codex prompt
on a real project folder.

See you in Lecture 5 !"

'''
```
